```python
# Wordcloud stuff comes from here:
# https://github.com/amueller/word_cloud

import matplotlib.pyplot as plt
import wordcloud
```

```python
def read_file(name):
    with file('data/' + name, 'r') as handle:
        return handle.read()
```

```python
raw = read_file(files[1])
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = (10,10)
```

```python
def make_word_cloud(text, stops=None):
    import wordcloud
    # you have to figure out a font path for your machine (sigh)
    myWordcloud = wordcloud.WordCloud(font_path="/Users/lynn/Library/Fonts/Lato-Medium.ttf", stopwords=stops).generate(text)
    plt.figure(figsize=(9,9))
    plt.imshow(myWordcloud)
    plt.axis("off")
    plt.show()
```

```python
make_word_cloud(raw)
```

```python
# Lesson1: inspect your algorithm and garbage in, garbage out.  Wordcloud also allows passing in a custom 
# stopwords list (it just uses it's own otherwise); may want to join yours with the english one and use those.
# Also, it does its own tokenization! so give it something you already cleaned a bit.
```

```python
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from IPython.display import clear_output, display, HTML


def make_word_cloud_topics(topic=None, stops=None):
    from nltk.corpus import stopwords
    english_stops = set(stopwords.words('english'))
    def make_pic(topic, stops):
        stops = stops.split() # it's a string of words spaced
        newstops = english_stops
        newstops.update(stops) # beware, we can't really go back now - once it's been updated
        text = read_file(topic)
        make_word_cloud(text, stops=newstops)
    interact(make_pic, topic = files, stops='')
```

```python
# this lets you add stopwords separated by spaces, live... regular english stops are already filtered out.
make_word_cloud_topics()
```
